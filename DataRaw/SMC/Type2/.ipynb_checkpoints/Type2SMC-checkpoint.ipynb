```python
fname1='../../SMC/Type1/LIN178/ADP.2023-01-09T23:22:14.152.fits'
fname2='../../SMC/Type1/LIN360/ADP.2020-08-14T15:43:54.025.fits'
fname3='../../SMC/Type1/OGLEJ005312.58-725533.6/ADP.2022-09-26T05:41:40.280.fits'
fname4='../../SMC/Type1/OGLEJ010838.56-723633.5/ADP.2014-10-07T08:35:09.440.fits'
mu_cent='../../SMC/Mu_Centauri/ADP.2016-09-21T11:56:32.916.fits'
```

```python
#plt.figure(figsize=(20,20))
#plt.imshow(fits.getdata(fname3).T[600:800],norm='log',vmin=200,aspect=10)
xd1 = fits.getdata(fname1)
xd2 = fits.getdata(fname2)
xd3 = fits.getdata(fname3)
xd4 = fits.getdata(fname4)
cent=fits.getdata(mu_cent)
xd2,cent
```

```python
fig, axs = plt.subplots(4, 1, sharex=True, figsize=(8, 5))  # 5 rows, 1 column

# Plot each light curve
axs[0].scatter(xd1['WAVE'][0]-4.5,xd1['FLUX'][0],s=0.1, color='black', label='OGLEJ045958.36-692549.4')
axs[0].scatter(cent['WAVE'][0],cent['FLUX'][0]-3,s=0.1, color='red')
axs[1].scatter(xd2['WAVE'][0]-4.5,xd2['FLUX_REDUCED'][0], s=0.1, color='black',label='OGLEJ050646.31-6843094')
axs[1].scatter(cent['WAVE'][0],cent['FLUX'][0]+950,s=0.1, color='red')
axs[2].scatter(xd3['WAVE'][0]-4.5,xd3['FLUX'][0], s=0.1, color='black',label='OGLEJ050650.94-700052.7')
axs[2].scatter(cent['WAVE'][0],cent['FLUX'][0]+18,s=0.1, color='red')
axs[3].scatter(xd4['WAVE'][0]-4.5,xd4['FLUX'][0], s=0.1, color='black',label='OGLEJ050652.74-683437.4')
axs[3].scatter(cent['WAVE'][0],cent['FLUX'][0]+40,s=0.1, color='red')



# Labels
for ax in axs:
    ax.set_ylabel('Amplitud') 
    ax.axvline(x=6562.8, color='blue', linestyle='--', linewidth=1)
    ax.legend(loc='lower right', fontsize='x-small')
axs[-1].set_xlabel(r'Longitud de Onda [$\AA$]')
axs[-1].set_xlim(6440, 6680)
axs[1].set_ylim(400, 1100)

fig.suptitle('LMC Type 1')
plt.tight_layout()
plt.show()
```
