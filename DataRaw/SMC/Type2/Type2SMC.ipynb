```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.modeling import models, fitting
fitter = fitting.LinearLSQFitter()
from scipy.ndimage import median_filter
from sklearn.preprocessing import MinMaxScaler
```

```python
fname1='../../SMC/Type2/2dFS440/ADP.2022-09-26T05:42:50.915.fits'
fname2='../../SMC/Type2/2dFS471/ADP.2014-10-07T13:37:18.583.fits'
fname4='../../SMC/Type2/MA93167/ADP.2016-09-21T09:18:32.409.fits'
fname3='../../SMC/Type2/OGLEJ004319.46-731705.7/ADP.2014-09-26T16:52:09.357.fits'
mu_cent='../../SMC/Mu_Centauri/ADP.2016-09-21T11:56:32.916.fits'
```

```python
scaler = MinMaxScaler()
#fits.getheader(mu_cent)
```

```python
fits.info(fname1)
```

```python
#plt.figure(figsize=(20,20))
#plt.imshow(fits.getdata(fname3).T[600:800],norm='log',vmin=200,aspect=10)
xd1 = fits.getdata(fname1)
xd2 = fits.getdata(fname2)
xd3 = fits.getdata(fname3)
xd4 = fits.getdata(fname4)
cent=fits.getdata(mu_cent)
xd2,cent
```

```python
#star 1
xd1_new=xd1['FLUX_REDUCED'][0][4145:-1719].tolist()
xd1_filt=median_filter(xd1_new,50)
lam1 = xd1['WAVE'][0][4145:-1719]

mask_alpha1=np.abs(6562.8-lam1)>20
mask_beta1=np.abs(4861.33333333333-lam1)>40
mask_gamma1=np.abs(4367.23-lam1)>50 #H_gamma est치 corrido
mask1=mask_alpha1&mask_beta1&mask_gamma1

#star 2
xd2_new=xd2['FLUX'][0][4145:-1719].tolist()
xd2_filt=median_filter(xd2_new,250)
lam2 = xd2['WAVE'][0][4145:-1719]

mask_alpha2=np.abs(6562.8-lam2)>40
mask_beta2=np.abs(4861.33333333333-lam2)>40
mask_gamma2=np.abs(4367.23-lam2)>50 #H_gamma est치 corrido
mask2=mask_alpha2&mask_beta2&mask_gamma2

#star 3
xd3_new=xd3['FLUX'][0][4145:-1719].tolist()
xd3_filt=median_filter(xd3_new,250)
lam3 = xd3['WAVE'][0][4145:-1719]

mask_alpha3=np.abs(6562.8-lam3)>40
mask_beta3=np.abs(4861.33333333333-lam3)>40
mask_gamma3=np.abs(4367.23-lam3)>50 #H_gamma est치 corrido
mask3=mask_alpha3&mask_beta3&mask_gamma3

#MuCen
cent_new=cent['FLUX'][0][4145:-1719].tolist()
cent_filt=median_filter(cent_new,250)
lamc = cent['WAVE'][0][4145:-1719]

mask_alphac=np.abs(6562.8-lamc)>40
mask_betac=np.abs(4861.33333333333-lamc)>40
mask_gammac=np.abs(4367.23-lamc)>50 #H_gamma est치 corrido
maskc=mask_alphac&mask_betac&mask_gammac
```

```python
modelo1 = fitter(models.Chebyshev1D(25),lam1[mask1],xd1_filt[mask1])
modelo2 = fitter(models.Chebyshev1D(25),lam2[mask2],xd2_filt[mask2])
modelo3 = fitter(models.Chebyshev1D(25),lam3[mask3],xd3_filt[mask3])
modeloc = fitter(models.Chebyshev1D(25),lamc[maskc],cent_filt[maskc])
```

```python
fig, axs = plt.subplots(3, 1, sharex=True, figsize=(8, 5))  # 5 rows, 1 column

# Plot each light curve
#axs[0].scatter(xd1['WAVE'][0]-2.5,xd1['FLUX'][0],s=0.1, color='black', label='OGLEJ004258.05-73023.7')
#axs[0].scatter(cent['WAVE'][0],cent['FLUX'][0]+20,s=0.1, color='red')
#axs[1].scatter(xd2['WAVE'][0]-4.5,xd2['FLUX'][0], s=0.1, color='black',label='OGLEJ004330.69-732034.2')
#axs[1].scatter(cent['WAVE'][0],cent['FLUX'][0]+60,s=0.1, color='red')
#axs[2].scatter(xd4['WAVE'][0]-4.5,xd4['FLUX'][0], s=0.1, color='black',label='OGLEJ004319.46-731705.7')
#axs[2].scatter(cent['WAVE'][0],cent['FLUX'][0]+40,s=0.1, color='red')

axs[0].scatter(lam1-3,xd1_filt/modelo1(lam1),s=0.1, color='black', label='OGLEJ004258.05-73023.7')
axs[0].scatter(lamc,(cent_filt/modeloc(lamc)),s=0.1, color='red',label=r'$\mu$-Cen')
axs[1].scatter(lam2-3,xd2_filt/modelo2(lam2), s=0.1, color='black',label='OGLEJ004330.69-732034.2')
axs[1].scatter(lamc,(cent_filt/modeloc(lamc)),s=0.1, color='red',label=r'$\mu$-Cen')
axs[2].scatter(lam3-3,xd3_filt/modelo3(lam3), s=0.1, color='black',label='OGLEJ004319.46-731705.7')
axs[2].scatter(lamc,(cent_filt/modeloc(lamc)),s=0.1, color='red',label=r'$\mu$-Cen')



# Labels
for ax in axs:
    ax.set_ylabel('Amplitud') 
    ax.axvline(x=6562.8, color='blue', linestyle='--', linewidth=1)
    ax.legend(loc='lower right', fontsize='x-small')
    ax.grid(True)

axs[-1].set_xlabel(r'Longitud de Onda [$\AA$]')
axs[-1].set_xlim(6440, 6680)
#axs[0].set_ylim(3, 32)
axs[1].set_ylim(0, 2)
axs[2].set_ylim(0, 2)

fig.suptitle('SMC Type 2')
plt.tight_layout()
plt.savefig('SMC_Type2')
plt.show()
```
