```python
fname1='../../LMC/Type2/OGLEJ050014.01-692739.6/ADP.2020-11-16T18:05:28.383.fits'
fname2='../../LMC/Type2/OGLEJ052407.54-695123.1/ADP.2020-10-20T14:11:37.497.fits'
fname3='../../LMC/Type2/UCAC21797339/ADP.2016-09-21T06:49:35.172.fits'
mu_cent='../../LMC/Mu_Centauri/ADP.2016-09-21T11:56:32.916.fits'
```

```python
#plt.figure(figsize=(20,20))
#plt.imshow(fits.getdata(fname3).T[600:800],norm='log',vmin=200,aspect=10)
xd1 = fits.getdata(fname1)
xd2 = fits.getdata(fname2)
xd3 = fits.getdata(fname3)
cent=fits.getdata(mu_cent)
xd1,cent
```

```python
fig, axs = plt.subplots(3, 1, sharex=True, figsize=(8, 5))  # 5 rows, 1 column

# Plot each light curve
axs[0].scatter(xd1['WAVE'][0]-4.5,xd1['FLUX'][0],s=0.1, color='black', label='OGLEJ050014.01-692739.6')
axs[0].scatter(cent['WAVE'][0],cent['FLUX'][0]+150,s=0.1, color='red')
axs[1].scatter(xd2['WAVE'][0]-4.5,xd2['FLUX'][0], s=0.1, color='black',label='OGLEJ052407.54-695123.1')
axs[1].scatter(cent['WAVE'][0],cent['FLUX'][0],s=0.1, color='red')
axs[2].scatter(xd3['WAVE'][0]-4.5,xd3['FLUX'][0], s=0.1, color='black',label='OLGEJ050346.47-690201.9')
axs[2].scatter(cent['WAVE'][0],cent['FLUX'][0],s=0.1, color='red')



# Labels
for ax in axs:
    ax.set_ylabel('Amplitud') 
    ax.axvline(x=6562.8, color='blue', linestyle='--', linewidth=1)
    ax.legend(loc='lower right', fontsize='x-small')
axs[-1].set_xlabel(r'Longitud de Onda [$\AA$]')
axs[-1].set_xlim(6440, 6680)
plt.title('LMC Type 2')
plt.tight_layout()
plt.show()
```
